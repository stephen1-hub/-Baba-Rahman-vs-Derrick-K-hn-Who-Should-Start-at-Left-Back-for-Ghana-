# Create Per-90 Metrics

```python
per90_cols = [
    "Goals",
    "Assists",
    "Shots",
    "Chances_Created",
    "Successful_Crosses",
    "Successful_Dribbles",
    "Duels_Won",
    "Tackles",
    "Interceptions",
    "Recoveries",
    "Clearances"
]

for col in per90_cols:
    df[f"{col}_Per90"] = df[col] / df["90s"]
```

# Create Category Scores

```python
df["Attack_Score"] = (
    df["Goals_Per90"] * 4 +
    df["Assists_Per90"] * 3 +
    df["xA"] * 2 +
    df["Chances_Created_Per90"]
)


df["Creativity_Score"] = (
    df["xA"] * 3 +
    df["Chances_Created_Per90"] * 3 +
    df["Successful_Crosses_Per90"] * 2
)


df["Defense_Score"] = (
    df["Tackles_Per90"] +
    df["Interceptions_Per90"] +
    df["Recoveries_Per90"] +
    df["Clearances_Per90"]
)

df["Possession_Score"] = (
    df["Pass_Accuracy"] +
    df["Duel_Win_Pct"] +
    df["Aerial_Win_Pct"]
)

df["Discipline_Score"] = (
    100 -
    (df["Yellow_Cards"] * 5) -
    (df["Red_Cards"] * 20)
)
```

# Build Radar Scores From Your Data

# Attack

```python
df["Attack_Radar"] = (
    df["Attack_Score"] /
    df["Attack_Score"].max()
) * 100
```

# Creativity

```python
df["Creativity_Radar"] = (
    df["Creativity_Score"] /
    df["Creativity_Score"].max()
) * 100
```

# Defense

```python
df["Defense_Radar"] = (
    df["Defense_Score"] /
    df["Defense_Score"].max()
) * 100
```

# Possession

```python
df["Possession_Radar"] = (
    df["Possession_Score"] /
    df["Possession_Score"].max()
) * 100
```

# Discipline

```python
df["Discipline_Radar"] = (
    df["Discipline_Score"] /
    df["Discipline_Score"].max()
) * 100
```

# Create Final Readiness Index

```python
radar = pd.DataFrame({
    "Metric":[
        "Attack",
        "Creativity",
        "Defense",
        "Possession",
        "Discipline"
    ],

    "Baba":[
        df.loc[0,"Attack_Radar"],
        df.loc[0,"Creativity_Radar"],
        df.loc[0,"Defense_Radar"],
        df.loc[0,"Possession_Radar"],
        df.loc[0,"Discipline_Radar"]
    ],

    "Kohn":[
        df.loc[1,"Attack_Radar"],
        df.loc[1,"Creativity_Radar"],
        df.loc[1,"Defense_Radar"],
        df.loc[1,"Possession_Radar"],
        df.loc[1,"Discipline_Radar"]
    ]
})
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=radar["Baba"],
    theta=radar["Metric"],
    fill='toself',
    name='Baba Rahman'
))

fig.add_trace(go.Scatterpolar(
    r=radar["Kohn"],
    theta=radar["Metric"],
    fill='toself',
    name='Derrick Köhn'
))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0,100]
        )
    ),
    showlegend=True
)

fig.show()
```
